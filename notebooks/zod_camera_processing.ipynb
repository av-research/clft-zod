# Camera Processing with Annotations

### Annotation Types
- **Object Detection**: Bounding boxes with color coding
- **Lane Markings**: Semi-transparent yellow overlay
- **Ego Road**: Semi-transparent purple overlay  
- **OXTS Data**: GPS/INS sensor visualization

## Annotation Color Coding

The images above show various annotation types with the following color scheme:

### Object Detection Bounding Boxes
- üî¥ **Vehicle**: Dark red `(100, 0, 0)` - Cars, trucks, buses
- üü¢ **Pedestrian**: Dark green `(0, 100, 0)` - People walking
- üîµ **TrafficSign**: Dark blue `(0, 0, 100)` - Traffic signs and signals
- üü¶ **VulnerableVehicle**: Dark cyan `(0, 100, 100)` - Bicycles, motorcycles
- ‚ö´ **Other**: Black `(0, 0, 0)` - Unspecified objects

### Overlay Annotations
- üü° **Lane Markings**: Semi-transparent yellow `(100, 100, 0, Œ±=0.5)`
- üü£ **Ego Road**: Semi-transparent purple `(100, 0, 100, Œ±=0.5)`
- üìç **OXTS Data**: GPS/INS sensor information (various colors)

## Example Output Images

Here's how to display example annotated images from the output folder:

```python
# Display example annotated images from output folder
import matplotlib.pyplot as plt
from PIL import Image
import os

# Select example images to display
example_files = [
    "camera_009158.png",
    "camera_018591.png", 
    "camera_062592.png",
    "camera_083430.png"
]

output_dir = "../output/camera"

# Create a 2x2 grid for displaying images
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.ravel()

for i, filename in enumerate(example_files):
    filepath = os.path.join(output_dir, filename)
    
    if os.path.exists(filepath):
        try:
            # Load and display image
            img = Image.open(filepath)
            axes[i].imshow(img)
            axes[i].set_title(f'{filename}', fontsize=14, fontweight='bold')
            axes[i].axis('off')
            print(f"‚úÖ Loaded: {filename}")
        except Exception as e:
            print(f"‚ùå Error loading {filename}: {e}")
    else:
        print(f"‚ö†Ô∏è File not found: {filename}")

plt.tight_layout()
plt.show()
```
